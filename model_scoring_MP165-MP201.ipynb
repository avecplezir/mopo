```python
datasets = [
    "D3RLPY-MP1",
    "D3RLPY-MP1_100000",
    "D3RLPY-MP1-P0-3",
    "D3RLPY-MP1-P1-4",
    "D3RLPY-MP1-P0_25000",
    "D3RLPY-MP1-P1_25000",
    "D3RLPY-MP1-P2_25000",
    "D3RLPY-MP1-P3_25000",
    "D3RLPY-MP1-P4_25000",
    "D3RLPY-MP1-P0_100000",
    "D3RLPY-MP1-P1_100000",
    "D3RLPY-MP1-P2_100000",
    "D3RLPY-MP1-P3_100000",
    "D3RLPY-MP1-P4_100000",
    "D3RLPY-MP2-P0_100000",
    "D3RLPY-MP2-P1_100000",
    "D3RLPY-MP2-P2_100000",
    "D3RLPY-MP2-P3_100000",
    "D3RLPY-MP2-P4_100000",
    "D3RLPY-MP3-P0_100000",
    "D3RLPY-MP3-P1_100000",
    "D3RLPY-MP3-P2_100000",
    "D3RLPY-MP3-P3_100000",
    "D3RLPY-MP3-P4_100000",
    "D3RLPY-PAP5",
    "D3RLPY-PAP5_100000",
    "D3RLPY-PAP5-P0-3",
    "D3RLPY-PAP5-P1-4",
    "D3RLPY-PAP5-P0_25000",
    "D3RLPY-PAP5-P1_25000",
    "D3RLPY-PAP5-P2_25000",
    "D3RLPY-PAP5-P3_25000",
    "D3RLPY-PAP5-P4_25000",
    "D3RLPY-PAP5-P0_100000",
    "D3RLPY-PAP5-P1_100000",
    "D3RLPY-PAP5-P2_100000",
    "D3RLPY-PAP5-P3_100000",
    "D3RLPY-PAP5-P4_100000",
    "D3RLPY-PAP6-P0_100000",
    "D3RLPY-PAP6-P1_100000",
    "D3RLPY-PAP6-P2_100000",
    "D3RLPY-PAP6-P3_100000",
    "D3RLPY-PAP6-P4_100000",
    "D3RLPY-PAP7-P0_100000",
    "D3RLPY-PAP7-P1_100000",
    "D3RLPY-PAP7-P2_100000",
    "D3RLPY-PAP7-P3_100000",
    "D3RLPY-PAP7-P4_100000",
    "RAND-1",
    "RAND-2",
    "RAND-3",
    "RAND-4",
    "RAND-5",
    "RAND-6",
    "RAND-7",
    "RAND-8",
    "RAND-9",
    "RAND-10",
]

d3rlpy_pap5_training_datasets = [
    "D3RLPY-PAP5-P0_25000",
    "D3RLPY-PAP5-P1_25000",
    "D3RLPY-PAP5-P2_25000",
    "D3RLPY-PAP5-P3_25000",
]

experiments_original = ['MP101', 'MP105', 'MP109', 'MP134', 'MP135', 'MP136']
experiments_mse_01  = ['MP165', 'MP166', 'MP167', 'MP168', 'MP169', 'MP170']
experiments_mse_1  = ['MP165', 'MP166', 'MP167', 'MP171', 'MP172', 'MP173']
experiments_mse_10  = ['MP165', 'MP166', 'MP167', 'MP174', 'MP175', 'MP176']
experiments_mse_tot_10  = ['MP199', 'MP200', 'MP201', 'MP183', 'MP184', 'MP185']
experiments_mse_tot_1  = ['MP199', 'MP200', 'MP201', 'MP186', 'MP187', 'MP188']
experiments_ll       = ['MP193', 'MP194', 'MP195', 'MP196', 'MP197', 'MP198']
```

```python
scores_original_df = get_scores_df(experiments_original, datasets)
scores_mse_01_df = get_scores_df(experiments_mse_01, datasets)
scores_mse_1_df = get_scores_df(experiments_mse_1, datasets)
scores_mse_10_df = get_scores_df(experiments_mse_10, datasets)
scores_mse_tot_1_df = get_scores_df(experiments_mse_tot_1, datasets)
scores_mse_tot_10_df = get_scores_df(experiments_mse_tot_10, datasets)
scores_ll_df = get_scores_df(experiments_ll, datasets)
```

```python
scores_original_groups = (
    scores_original_df.groupby(['training_dataset', 'rex', 'rex_beta', 'evaluation_dataset'])
)
scores_original_avgs_df = scores_original_groups.mean().reset_index()

scores_mse_01_groups = (
    scores_mse_01_df.groupby(['training_dataset', 'rex', 'rex_beta', 'evaluation_dataset'])
)
scores_mse_01_avgs_df = scores_mse_01_groups.mean().reset_index()

scores_mse_1_groups = (
    scores_mse_1_df.groupby(['training_dataset', 'rex', 'rex_beta', 'evaluation_dataset'])
)
scores_mse_1_avgs_df = scores_mse_1_groups.mean().reset_index()

scores_mse_10_groups = (
    scores_mse_10_df.groupby(['training_dataset', 'rex', 'rex_beta', 'evaluation_dataset'])
)
scores_mse_10_avgs_df = scores_mse_10_groups.mean().reset_index()

scores_mse_tot_1_groups = (
    scores_mse_tot_1_df.groupby(['training_dataset', 'rex', 'rex_beta', 'evaluation_dataset'])
)
scores_mse_tot_1_avgs_df = scores_mse_tot_1_groups.mean().reset_index()

scores_mse_tot_10_groups = (
    scores_mse_tot_10_df.groupby(['training_dataset', 'rex', 'rex_beta', 'evaluation_dataset'])
)
scores_mse_tot_10_avgs_df = scores_mse_tot_10_groups.mean().reset_index()

scores_ll_groups = (
    scores_ll_df.groupby(['training_dataset', 'rex', 'rex_beta', 'evaluation_dataset'])
)
scores_ll_avgs_df = scores_ll_groups.mean().reset_index()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(30, 10))

lss =  ['-', '--']
labels = ['Original', 'MSE Orig - Beta 0.1', 'MSE Orig - Beta 1.0', 'MSE Orig - Beta 10.0', 'MSE Tot - Beta 1.0', 'MSE Tot - Beta 10.0', 'LL']

metric = 'overall_mse'
for i, exp_set in enumerate((scores_original_avgs_df, scores_mse_01_avgs_df, scores_mse_1_avgs_df, scores_mse_10_avgs_df, scores_mse_tot_1_avgs_df, scores_mse_tot_10_avgs_df, scores_ll_avgs_df)):
    for j, rex in enumerate([False, True]):
        mask = (
            (exp_set['rex'] == rex)
        )
        plot_vals = exp_set.loc[mask].set_index('evaluation_dataset').loc[datasets][metric]
        p = ax.plot(
            datasets, plot_vals,
            label=f'{labels[i]} - REx: {rex}',
            ls=lss[j],
            c=cols[i]
        )
ax.set_xticks(ax.get_xticks(), datasets, rotation=80)
ax.set_xlabel('Evaluation Dataset')
ax.set_ylabel('Overall MSE')
ax.legend()
```
